```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
#loading the data
train_df=pd.read_csv('/content/train.csv')
test_df=pd.read_csv('/content/test.csv')
sample_df=pd.read_csv('/content/train.csv')
```

```python
#missing value handling
numeric_cols_train=train_df.select_dtypes(include=['float64','int64']).columns
numeric_cols_test=test_df.select_dtypes(include=['float64','int64']).columns
train_df[numeric_cols_train]=train_df[numeric_cols_train].fillna(train_df[numeric_cols_train].mean())
test_df[numeric_cols_test]=test_df[numeric_cols_test].fillna(test_df[numeric_cols_test].mean())
```

```python
#preprocessing the data
X=train_df.drop(columns=['ID','Target'])
y=train_df['Target']
```

```python
#standradize the features
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
test_scaled=scaler.transform(test_df.drop(columns=['ID']))

#split the traing data for validation
X_train,X_val,y_train,y_val=train_test_split(X_scaled,y,test_size=0.2,random_state=42)
```

```python
# train model
model=RandomForestClassifier(random_state=42)
model.fit(X_train ,y_train)
```

```python
#Validate the model
y_val_pred=model.predict(X_val)
accuracy=accuracy_score(y_val,y_val_pred)
print(f'Validation Accuracy:{accuracy*100:0.2f}% ')
```

```python
#predict the set
test_predictions=model.predict(test_scaled)
```

```python
#submission dataframe
submission_df=pd.DataFrame({'ID':test_df['ID'],'Target':test_predictions})
submission_df=submission_df.sort_values(by='ID')
submission_df.to_csv('submission.csv',index=False)
```
